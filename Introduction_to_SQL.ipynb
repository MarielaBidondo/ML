### FUNCTIONS

**4 MAX and MIN**

The MAX () function is used to extract the highest value from a data set. The MIN () function is used to extract data from the lowest value of a data set.


```
%%sql SELECT MIN (Price) AS SmallestPrice
FROM Products;
```

```
%%sql SELECT MAX (Price) AS LargestPrice
FROM Products;
```

**5 COUNT, AVG and SUM**

The COUNT () function returns the number of rows in a data set.
The AVG () function returns the mean of a numeric column from a data set.
The SUM () function returns the sum of a numeric column in a data set.


```
%%sql SELECT SUM(Quantity)
FROM order_details;
```

```
%%sql SELECT SUM (Quantity)
FROM order_details WHERE
OrderID = 10248;
```

**6 GROUP BY**

The GROUP BY statement is often used with aggregate functions (COUNT,
MAX, MIN, SUM, AVG) to group the result set by one or more
columns.


```
%%sql SELECT COUNT (CustomerID), Country
FROM Customers
GROUP BY Country;

```

```
%%sql SELECT COUNT (CustomerID), City,
Country FROM Customers
GROUP BY City
ORDER BY COUNT (CustomerID) DESC;
```

**7 CASE WHEN**

The CASE statement goes through conditions and they return a value when the
first condition (as an IF-THEN-ELSE statement). Thus, once the condition is true, it will stop reading and return the result. If the conditions are not true, returns the value of the ELSE clause.
If there are no other parts and they have no true conditions, it returns NULL.


```
%%sql SELECT OrderID, Quantity, CASE WHEN Quantity > 30 
THEN "The quantity is greater than 30" WHEN Quantity = 30 
THEN "The quantity is 30" ELSE "The quantity is under 30" 
END AS CLAUSE  FROM order_details LIMIT 10
```

```
%sql SELECT ProductName, IFNULL(SupplierID,5) FROM Products LIMIT 10
```

**8 UNION**

The UNION operator is used to combine the results of two or more SELECT statements.

● Each SELECT statement within union must have the same number of columns.

● The columns must also have similar data types.

● The columns of each SELECT statement must also be in the same order.

The UNION operator only selects different default values. To allow duplicate values,
use UNION ALL


```
%%sql SELECT City FROM Customers
UNION 
SELECT City FROM Suppliers LIMIT 10
```

```
%%sql SELECT City FROM Customers
UNION ALL
SELECT City FROM Suppliers LIMIT 10
```

```
%%sql SELECT City, Country FROM Customers WHERE Country = 'Germany'
UNION ALL SELECT City, Country 
FROM Suppliers WHERE Country = 'Germany' ORDER BY City;
```

**9 HAVING**

The HAVING clause was added because the WHERE clause could not be used with aggregate functions.


```
%%sql SELECT COUNT (CustomerID), Country
FROM Customers
GROUP BY Country
HAVING COUNT (CustomerID)>1;
```

```
%%sql SELECT COUNT (CustomerID) AS
COUNT, Country FROM Customers GROUP BY Country
HAVING COUNT (CustomerID)> 0
ORDER BY COUNT DESC;
```

## JOINS

**10 JOIN**

A JOIN clause is used to combine the rows of two or more tables, based on a related column between them.


**11 INNER JOIN**

The INNER JOIN keyword selects records that have matching values in both boards.




```
%%sql SELECT orders.OrderID,
customers.CustomerName
FROM orders
INNER JOIN customers ON
orders.CustomerID =
customers.CustomerID LIMIT 10
```

**12 LEFT JOINS**

The LEFT JOIN keyword returns all the records in the left table (table1), and the
matching records from the right table (table2). The result is NULL from the right side, if
there is no match.


```
%%sql SELECT orders.OrderID, customers.
CustomerName FROM orders
LEFT JOIN customers ON
orders.CustomerID =
customers.CustomerID LIMIT 10
```

**15 SELF JOINS**

A SELF JOIN is a regular JOIN, but the table is joined to itself.
Joining a table to itself means that each row in the table is combined with itself and with each
another row in the table.
It can be seen as a combination of two copies of the same table.

```
%%sql SELECT T1.Emp_ID AS Employee_ID, T1.Emp_Name AS Employee_Name, T2.Emp_ID AS Supervisor_ID, 
T2.Emp_Name AS Supervisor_Name FROM employees T1, employees T2 WHERE T1.Emp_Supv = T2.Emp_ID;
```
