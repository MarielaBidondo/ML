### Primera Pregunta:
- Cuáles son los géneros más alquilados y lo que menos

```python
%%sql

SELECT c.name AS Genero, count(r.rental_id) AS Total_Rentas
FROM rental AS r
JOIN inventory AS inv ON r.inventory_id = inv.inventory_id
JOIN film AS fil ON fil.film_id = inv.film_id 
JOIN film_category AS filca ON filca.film_id = fil.film_id
JOIN category AS c ON c.category_id = filca.category_id
GROUP BY c.name
ORDER BY 2 desc;
```

Uniendo las tablas de inventory, rental, film, film_category y category 
podemos conocer los generos, los agrupamos por nombre de genero
los ordenamos en orden descendente respecto a la segunda columna (la cantidad de alquileres)

Genero mas alquilado es Sport y el menos Music


### Segunda Pregunta:
- Cúales son los los géneros que más vendes y los que menos en términos de cantidad de dinero, mirad tabla de Payments

```python
%%sql

SELECT c.name AS Genero, SUM(p.amount) AS Ventas_totales
FROM payment AS p
JOIN rental AS r ON r.rental_id = p.rental_id
JOIN inventory AS inv ON inv.inventory_id = r.inventory_id
JOIN film AS fil ON fil.film_id = inv.film_id 
JOIN film_category AS filca ON filca.film_id = fil.film_id
JOIN category AS c ON c.category_id = filca.category_id
GROUP BY c.name
ORDER BY 2 desc;
```

Uniendo las tablas de payments, inventory, rental, film, film_category y category 
podemos conocer los generos, los agrupamos por nombre de genero
y sumamos los pagos, los ordenamos en orden descendente respecto a la segunda columna (la cantidad de dinero)

Generos que mas vende Sport y Sci-Fi  y los que menos menos Music y Travel


### Tercera Pregunta:
- Cuantos usuarios únicos han alquilado cada género?

```python
%%sql

SELECT c.name AS Genero, COUNT(DISTINCT cu.customer_id) AS Usuarios_pG
FROM customer AS cu
JOIN rental AS r ON r.customer_id = cu.customer_id
JOIN inventory AS inv ON inv.inventory_id = r.inventory_id
JOIN film AS fil ON fil.film_id = inv.film_id 
JOIN film_category AS filca ON filca.film_id = fil.film_id
JOIN category AS c ON c.category_id = filca.category_id
GROUP BY c.name
ORDER BY 2 desc;
```

Uniendo las tablas de customer, inventory, rental, film, film_category y category 
podemos conocer los generos, los agrupamos por nombre de genero. Para identificar usuarios unicos usamos DISTINCT, los ordenamos en orden descendente respecto a la segunda columna (la suma de usuarios unicos que han rentado ese genero)

Generos que mas alquilados por usuarios unicos Sports - Action y los menoos travel y music

### Cuarta Pregunta:
- Cuanto es el gasto medio en alquiler por cada cliente?

```python
%%sql

SELECT cu.customer_id, cu.first_name AS Nombre, cu.last_name AS Apellido,
ROUND(AVG(fil.rental_rate),2) AS Gasto_Medio
FROM film AS fil 
JOIN inventory AS inv ON inv.film_id = fil.film_id 
JOIN rental AS r ON inv.inventory_id = r.inventory_id
JOIN customer AS cu ON r.customer_id = cu.customer_id
GROUP BY cu.customer_id
ORDER BY 4 desc
LIMIT 10;
```

Uniendo las tablas de customer, inventory, rental, film podemos conocer los pagos por film, los agrupamos por usuarios. Para identificar usuarios unicos usamos DISTINCT, los ordenamos en orden descendente respecto a la segunda columna (Quien gasta mas o menos en promedio).Para simplificar la tabla podemmos usar ROUND, de esta forma  

Porque son demasiados reusltados he limitado el query a solo a 10 resultado, ordenados por gasto promedio Sam, Mcduffie mayor gasto -> 3.82 y menor Leona, Obrien -> 1.85 

### Quinta Pregunta
- Ordenar los países por los que más alquilan y más clientes tienen

```python
%%sql

SELECT cou.country AS pais, COUNT(DISTINCT cu.customer_id) AS Clientes,
COUNT(rental_id) AS total_alquiler
FROM customer AS cu
JOIN rental AS r ON cu.customer_id = r.customer_id
JOIN address AS add ON cu.address_id = add.address_id
JOIN city AS ci ON add.city_id = ci.city_id
JOIN country AS cou ON ci.country_id = cou.country_id
GROUP BY cou.country
ORDER BY 3 desc, 2 desc;
```

Uniendo las tablas de customer, rental, address, city, country  podemos conocer los paises y cantidad de rentas, los agrupamos por pais.De los paises el que mas cliente y rentas tiene es India
Los ordenamos de mayor a menor en numero de rentas y luego en numero de clientes usando Order by desc

### Sexta Pregunta:
- El top 10 de clientes que más alquilan 

```python
%%sql

SELECT cu.customer_id, cu.first_name AS Nombre, cu.last_name AS Apellido,
COUNT(r.rental_id) AS total_alquiler
FROM film AS fil 
JOIN inventory AS inv ON inv.film_id = fil.film_id 
JOIN rental AS r ON inv.inventory_id = r.inventory_id
JOIN customer AS cu ON r.customer_id = cu.customer_id
GROUP BY cu.customer_id
ORDER BY 4 desc
LIMIT 10;
```

Uniendo las tablas de customer, inventory, rental, film podemos conocer las rentas, los agrupamos por usuarios. Para identificar usuarios que alquilaron mas, los ordenamos usando order by

Como necesitamos ver los primeros 10, limitaremos la tabla a 10 con limit

### Séptima Pregunta:
- Conteo de películas que han sido devueltas más tarde.(Pista: 
  tabla Rental, mirar la diferencia entre return_date y rental_date, y comparar con rental_duration de la tabla Film)

```python
%%sql

SELECT CASE 
WHEN rental_duration < DATE_PART('day', return_date - rental_date)THEN 'Devolucion tarde'
ELSE 'Devolucion en tiempo o processo'
END AS devolucion,
COUNT(*) AS numero_total
FROM film AS f
JOIN inventory AS inv ON f.film_id = inv.film_id
JOIN rental AS r ON inv.inventory_id = r.inventory_id
GROUP BY devolucion
ORDER BY 1 desc 
```

Con las tablas film, rental e inventry sacaremos la informacion
de el dia de alquiler, dia de retorno y el tiempo de alquiler establecido.
Usando la funcion DATE_PART() podemos calcular el tiempo en que el usuario tuvo el film.
Selecionamos la unidad 'day', ya que el tiempo de retorno esta estipulado en esa unidad.
Aplicaremos un CASE para distinguir de films que han sido devueltas mas tarde de lo esperado.
Podemos ver un total de 6403 peliculas devueltas tarde
